```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\josep\OneDrive\PYTHON PROJECT (CORRELATION)\movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# Checking if there is any missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Dropping all rows that have missing data

df = df.dropna()
```

```python
# Checking the data types of our columns

df.dtypes
```

```python
# Changing the datatype of some columns

df = df.fillna(0)
df[['budget', 'gross']] = df[['budget', 'gross']].astype('int64')

```

```python
df
```

```python
# Creating correct Year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Order the table by the gross column

df = df.sort_values(by=['gross'], inplace = False, ascending = False)

df
```

```python
# Creating a Scatter Plot: Budget vs Gross

plt.scatter(x=df['budget'],y=df['gross']) 

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget Earnings')

plt.show
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Look at correlation

df.corr(method = 'pearson')
```

```python
# Plotting Correlation Matrix

correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
# Numerizing all data

df_numerized = df

for col_name in df_numerized.columns:
        if (df_numerized[col_name].dtype == 'object'):
            df_numerized[col_name] = df_numerized[col_name].astype('category')
            df_numerized[col_name] = df_numerized[col_name].cat.codes
            
df_numerized
```

```python
df
```

```python
# Plotting Correlation Matrix for df_numerized

correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
df_numerized.corr()
```

```python
# Looking at correlation pairs

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Looking at correlation pairs in a different way

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Looking at correlation pairs with a correlation value greater than 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
